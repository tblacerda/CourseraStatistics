```python
import pandas as pd
```

```python
df_basic = pd.read_csv('data.tsv', sep='\t')
```

```python
df_basic['titleType'].value_counts()
```

```python
df_basic = df_basic.loc[df_basic['titleType']=='movie']
```

```python
df_basic = df_basic[df_basic['genres'].str.contains("War")]
```

```python
df_basic.to_excel('treated.xlsx')
```

```python
df_rating = pd.read_csv('basics.tsv', sep ='\t')
```

```python
df_merge = pd.merge(df_basic, df_rating, how='left', on='tconst')
```

```python
df_merge.dropna(axis=0,subset=['averageRating'],inplace=True)
```
